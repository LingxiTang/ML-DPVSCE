### This notebook takes in csv dataset of input features and output variable to perform the following hyperparameter optimisation:
- Number of hidden neurons
- Initial learning rate and alpha (L2 regularisation parameter)
- Maximum iterations

```python
## Import libraries 
from sklearn.neural_network import MLPRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split, validation_curve
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import math
from sklearn import preprocessing
import keras
from matplotlib.lines import Line2D
import time
from sklearn.model_selection import GridSearchCV
```

```python
# Import training set
df1 = pd.read_csv (r"C:\Users\tangl\OneDrive\Engineering Course\Year 4\4YP Lingxi Tang\Training Datasets\Datasetv4_ML_all_1.csv",index_col=[0])
df2 = pd.read_csv (r"C:\Users\tangl\OneDrive\Engineering Course\Year 4\4YP Lingxi Tang\Training Datasets\Datasetv4_ML_all_2.csv",index_col=[0])
df = pd.concat([df1,df2])

# Remove Peak_all features
df = df.drop(df.columns[[14,15,16,17,18,19,20]],axis = 1)
```

```python
X = df.iloc[:,:16]    # Change the number value according to number of input features: 16 for all seasons
Y = df.iloc[:,-1:]
X = X.to_numpy()
Y = Y.to_numpy()

(X_trainval, X_test, y_trainval, y_test,) = train_test_split(X, Y,stratify = Y, test_size=0.1,random_state = 1)
```

```python
## Standardisation of input features
sc_X = preprocessing.StandardScaler()
standardised_Xtrainval = sc_X.fit_transform(X_trainval)
y_trainval = np.ravel(y_trainval)
```

### Optimising number of hidden neurons

```python
## Choose number of hidden neurons to be tested
parameters = {'hidden_layer_sizes':[(x,x) for x in range(10,500,10)]}  
## Perform grid search 
nn = MLPRegressor()
clf = GridSearchCV(nn, parameters,scoring = 'neg_root_mean_squared_error')
clf.fit(standardised_Xtrainval,y_trainval)
```

```python
mean_fit_time = clf.cv_results_['mean_fit_time']
std_fit_time = clf.cv_results_['std_fit_time']
mean_test_score = clf.cv_results_['mean_test_score']
mean_test_score = np.negative(mean_test_score)
std_test_score = clf.cv_results_['std_test_score']
neurons = np.arange(10,500,10)
```

```python
## Plot RMSE and training duration against number of hidden neurons
# create figure and axis objects with subplots()
fig,ax = plt.subplots(figsize=(10,5))
# make a plot
ax.plot(neurons, mean_fit_time, color="red")
# set x-axis label
ax.set_xlabel("Hidden neurons",fontsize =15)
# set y-axis label
ax.set_ylabel("Training time (s)",color="red",fontsize =15)
# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(neurons, mean_test_score,color="blue")
ax2.set_ylabel("RMSE (kW)",color="blue",fontsize =15)

ax.fill_between(neurons, mean_fit_time-(std_fit_time/2), mean_fit_time+(std_fit_time/2),alpha = 0.3,color = "red")
ax2.fill_between(neurons, mean_test_score-(std_test_score/2), mean_test_score+(std_test_score/2),alpha = 0.3,color = "blue")


plt.show()
```

### Optimising learning rate and alpha

```python
## Choose set of initial learning rate and alpha values to be tested
parameters = {'learning_rate_init':(0.05,0.01,0.0075), 'alpha':(0.00075,0.0005,0.00025)}
## Perform grid search
nn = MLPRegressor(hidden_layer_sizes = (270,270))
clf = GridSearchCV(nn, parameters,scoring = 'neg_root_mean_squared_error')
clf.fit(standardised_Xtrainval,y_trainval)
```

```python
## Show best combination of alpha and initial learning rate
clf.best_params_
```

### Optimising maximum iterations

```python
## Max iterations to be changed at "param_range" parameter
train_scores, valid_scores = validation_curve(MLPRegressor(hidden_layer_sizes = (270,270), learning_rate_init = 0.01, alpha = 0.0005), standardised_Xtrainval, y_trainval, param_name="max_iter", param_range=np.arange(10,410,10), cv=5,scoring = 'neg_root_mean_squared_error')
```

```python
train_scores = np.negative(train_scores)
valid_scores = np.negative(valid_scores)
```

```python
## Plot training and validation RMSE against max iterations
x = np.arange(10,410,10)
plt.figure(figsize=(10,5) )

plt.plot(x,np.mean(valid_scores,axis = 1),label = 'Validation RMSE',marker = 'o',color = 'blue')
plt.plot(x,np.mean(train_scores,axis = 1),label = "Training RMSE",marker = 'o',color = 'red')
#plt.fill_between(x, np.mean(valid_scores,axis = 1)-(np.std(valid_scores,axis = 1)/2), np.mean(valid_scores,axis = 1)+(np.std(valid_scores,axis = 1)/2),alpha = 0.3,color = "blue")
#plt.fill_between(x, np.mean(train_scores,axis = 1)-(np.std(train_scores,axis = 1)/2), np.mean(train_scores,axis = 1)+(np.std(train_scores,axis = 1)/2),alpha = 0.3,color = "red")
plt.legend()
plt.xlabel("Max Iterations")
plt.ylabel("RMSE (kW)")
plt.show
```

### Plotting training time against maximum number of iterations

```python
parameters = {'max_iter':[x for x in np.arange(10,410,10)]}
nn = MLPRegressor(hidden_layer_sizes= (270,270),learning_rate_init=0.01,alpha = 0.0005)
clf = GridSearchCV(nn, parameters,scoring = 'neg_root_mean_squared_error')
clf.fit(standardised_Xtrainval,y_trainval)
mean_fit_time = clf.cv_results_['mean_fit_time']
std_fit_time = clf.cv_results_['std_fit_time']
max_iter = np.arange(10,410,10)# create figure and axis objects with subplots()
fig,ax = plt.subplots(figsize=(10,5))
# make a plot
ax.plot(max_iter, mean_fit_time, color="blue")
# set x-axis label
ax.set_xlabel("Max Iterations",fontsize =15)
# set y-axis label
ax.set_ylabel("Training time (s)",fontsize =15)


ax.fill_between(max_iter, mean_fit_time-(std_fit_time/2), mean_fit_time+(std_fit_time/2),alpha = 0.3,color = "blue")


plt.show()
```

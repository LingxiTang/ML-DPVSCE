## This notebook is used to generate Figure 3.5 in the report


```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def sigmoid(x):
    sig = 1 / (1 + math.exp(-x))
    return sig

def relu(x):
    relu = max(0,x)
    return relu
```

```python
x = np.arange(-2,2.1,0.1)
sigmoid_y = []

for i in range(len(x)):
    sigmoid_y.append(sigmoid(x[i]))
    
relu_y = []
for i in range(len(x)):
    relu_y.append(relu(x[i]))
```

```python
plt.plot(x, np.tanh(x),label = "Tanh function",color = "blue")
plt.plot(x,sigmoid_y,label = "Sigmoid function",color = "green")
plt.plot(x,relu_y,label = "ReLU function",color = "orange")
plt.ylim(-1.5,2.5)
plt.ylabel("Output")
plt.xlabel("Input")
plt.xticks((-2,-1,0,1,2))
plt.yticks((-2,-1,0,1,2))

plt.legend()
plt.grid()
```
